```python
#Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read dataset
datas=pd.read_csv("n_movies.csv")
datas
```

```python
#Find the shape of the dataset
print("The shape of the file (Rows, Columns)=",datas.shape)
```

```python
#information about dataset
print('The information about the data',datas.info())
```

```python
#describing basic analysis about dataset based on rating
#rating only has numeric value
datas.describe()
```

```python
datas.head()
```

```python
#list sort by top rating
TopRating=datas.sort_values('rating',ascending=False)
TopRating
```

```python
Top10=TopRating.head(20)
Top10
```

```python
#distinct top 10 titles
Top10['title'].value_counts()
```

```python
#distinct top genres
Top10['genre'].value_counts()
```

```python
#visual present for top 10 movies
plt.style.use('bmh')
plt.figure(figsize=(8,8))
plt.barh(Top10['title'],Top10['rating'])
plt.show()
```

```python
#visual present for top genre movies
Top10['genre'].value_counts().plot(kind='barh')
```
